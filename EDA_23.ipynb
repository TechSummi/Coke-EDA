To compare the current PO price("Net_Order_Price_Per_Unit") with the previous PO price("previous_Net_Order_Price_Per_Unit") for the same Material , Plnt and Vendor.

```python
# Create a new column with the previous PO price
db1_sorted['previous_Net_Order_Price_Per_Unit'] = db1_sorted.groupby(['Material', 'Plnt', 'Vendor'])['Net_Order_Price_Per_Unit'].shift()

# Compare the current PO price with the previous PO price
db1_sorted['PO_Price_Change'] = np.where(db1_sorted['Net_Order_Price_Per_Unit'] > db1_sorted['previous_Net_Order_Price_Per_Unit'], 'Increase',
                                        np.where(db1_sorted['Net_Order_Price_Per_Unit'] < db1_sorted['previous_Net_Order_Price_Per_Unit'], 'Decrease', 'No Change'))

# Display the updated dataframe
print(db1_sorted.head())  # display the first few rows of the updated dataframe
```

Made groups for the same Material , Plnt and Vendor 

```python
grouped = db1_sorted.groupby(['Material', 'Plnt','Vendor', 'PO Created on'])['Net_Order_Price_Per_Unit'].mean().reset_index()
```

```python
# Create a group number column in the original dataframe db1_sorted 
db1_sorted['Group Number'] = db1_sorted.groupby(['Material', 'Plnt', 'Vendor', 'PO Created on']).ngroup().add(1).fillna(-1).astype(int).astype(str).replace('-1', 'Not a group')

# Prefix the group number with 'G'
db1_sorted['Group Number'] = 'G' +  db1_sorted['Group Number'].replace('Not a group', 'Not a group')

db1_sorted['Group Number'] = np.where(db1_sorted['Group Number'] != 'GNot a group', 
                                      'G' + db1_sorted['Group Number'], 
                                      db1_sorted['Group Number'])
```

```python
# Iterate through each unique group
for group in db1_sorted['Group Number'].unique():
    # Filter the dataframe for the current group
    df_group = db1_sorted[db1_sorted['Group Number'] == group].copy()
```

To check if the "PO Created on" date difference between the current transaction and the last transaction is less than 90 days,
Calculate the price variation percentage.

```python
db1_sorted['PO Created on']=pd.to_datetime(db1_sorted['PO Created on'],format='mixed')
db1_sorted['Document Date']=pd.to_datetime(db1_sorted['Document Date'],format='mixed')
db1_sorted['Changed On']=pd.to_datetime(db1_sorted['Changed On'],format='mixed')
```

```python
db1_sorted['PO Created on'] = pd.to_datetime(db1_sorted['PO Created on'])
```

```python
# Calculate the date difference between the current and previous transactions
db1_sorted['Date Difference'] = (db1_sorted['PO Created on'] - db1_sorted.groupby(['Material', 'Plnt', 'Vendor'])['PO Created on'].shift()).dt.days

# Check if the date difference is less than 90 days
db1_sorted['Within 90 Days'] = np.where(db1_sorted['Date Difference'] < 90, True, False)

# Calculate the price variation percentage
db1_sorted['Price Variation %'] = ((db1_sorted['Net_Order_Price_Per_Unit'] - db1_sorted['previous_Net_Order_Price_Per_Unit']) / db1_sorted['previous_Net_Order_Price_Per_Unit']) * 100

# Display the updated dataframe
print(db1_sorted.head())
```

To identify anomalies if the price variation % is more than 10%

```python
# Identify anomalies if the price variation % is more than 10%
db1_sorted['Anomaly'] = np.where(db1_sorted['Price Variation %'] > 10, 'Yes', 'No')

print(db1_sorted.head())  
```
